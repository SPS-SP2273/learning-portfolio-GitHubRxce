## Yet another plot but with error bars

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', )
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
             label='Y3 with errors', color='green') # ADDING THE ERROR BAR FUNCTION
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

# Better with NumPy

Often,  it is easier to use NumPy arrays instead of Python lists.

```python
import numpy as np

# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
```

## Adding mathematical functions

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2                  # The math stuff
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend()                                 
```

## We need another axis!

Based on the above $sin(x)$ graph plotted, the plot does not look good because $sin(x)$ lies between $\pm1$, but $x$<sup>2</sup> has no such bounds. One way to fix this is to add another y-axis that shares the same x-axis by using the `plt.twinx()` function.

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2
```

### Things to note

In the $sin(x)$ graph above, we now have two `legend()` calls, one for each axis.

Our plot still does not look good because we have only a few points --> use `np.linspace` to fix this

```python
x = np.linspace(0, 6, 100) # USING np.linespace()
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')     

```

# Saving to disc

```python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.legend(loc='lower left')
plt.twinx()
plt.plot(x, sin_x,
         label='sin x', color='blue')
plt.legend(loc='lower right')

plt.savefig('simple-plot.png', dpi=150) # SAVING PLOT IN A DISK TO USE FOR REPORT OR PRESENTATION
                                        # STEP 1: Use 'plt.savefig()' function
                                        # STEP 2: Specify the format with the extension (e.g., filename.pdf) of the file name
                                        # STEP 3: Specify resolution by use dots-per-inch (dpi)
                                        # PLOT IS NOW SAVED IN SAME FILE AS THIS NOTEBOOK UNDER FILES
```

# A real example: Global Warming

## Plotting data from files

```python
data = np.loadtxt("C:\\Users\\USER\\Downloads\\Complete_TAVG_daily.txt",
                   skiprows=24)  # open/load the file using the file path and \\ to access the data, skip first 24 rows above the data table in the text doc

date = data[:, 0] # accessing the data from all rows ":", first column "0"
anomaly = data[:, -1] # accessing the data from all rows ":", last column "-1"

plt.plot(date, anomaly, alpha=.5) # used a small alpha value to soften the colour of the plot 
plt.ylim([-8, 8]) #  plot range symmetrical in the y direction
```

```python
data = np.loadtxt("C:\\Users\\USER\\Downloads\\Complete_TAVG_daily.txt",
                   skiprows=24)  

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])

plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey') # HIGHLIGHTING TREND 
                                                                # Use 'plt.hlines()'
                                                                # Include y-value, starting and ending x values, linestyle, and colors
```

## Styles

```python
plt.style.use('fivethirtyeight') # MUST BE THE VERY FIRST LINE OF CODE 

data = np.loadtxt("C:\\Users\\USER\\Downloads\\Complete_TAVG_daily.txt", skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.ylim([-8, 8])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
```
