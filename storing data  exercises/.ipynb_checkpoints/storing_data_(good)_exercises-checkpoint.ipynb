## Exercise 1

| Term       | Brief description                                    |
|------------|------------------------------------------------------|
| Subsetting | Selecting a subset of elements from a list/array/tuple             |
| Indexing   | Choosing _(selecting)_ an element from a list/array/tuple          |
| Slicing    | Choosing _(selecting)_ a range of elements from a list/array/tuple |
| Masking    | Masking False results by only showing True results   |

## Exercise 2

```python
%%timeit

import numpy as np

np_array_2d = np.array([[1, "A"], [3, "C"], [2, "B"], [4, "D"],
                        [5, "E"], [7, "G"], [6, "F"], [8, "H"],
                        [10, "J"], [9, "I"]])

# Subset all the first elements and convert to integers
numbers = np_array_2d[:, 0].astype(int)

# Use % to get the remainder for division by 2
remainder = numbers % 2

# Create a mask that checks if the remainder is zero or not, True = remainder not zero
mask = remainder != 0

# Use the mask to extract the corresponding second elements
result = np_array_2d[mask, 1]

print(result)
```

```python
%%timeit

list_2d = [[1, "A"], [3, "C"], [2, "B"], [4, "D"], [5, "E"], [7, "G"], [6, "F"], [8, "H"], [10, "J"], [9, "I"]]

# Subset all the first elements and convert to integers
numbers = [item[0] for item in list_2d]

# Use % to get the remainder for division by 2
remainder = [num % 2 for num in numbers]

# Create a mask that checks if the remainder is zero or not, True = remainder not zero
mask = [r != 0 for r in remainder]

# Use the mask to extract the corresponding second elements
result = [list_2d[i][1] for i in range(len(list_2d)) if mask[i]]

print(result)
```

Using list is faster than using numpy array
